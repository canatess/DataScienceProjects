Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
def happiness_degree_finder(column):
    if (column['Ladder']) >= 120:
        return 'very unhappy'
    elif (column['Ladder']) < 120 and (column['Ladder']) >= 90:
        return 'unhappy'
    elif (column['Ladder']) < 90 and (column['Ladder']) >= 60:
        return 'moderately happy'
    elif (column['Ladder']) < 60 and (column['Ladder']) >= 30:
        return 'happy'
    else:
        return 'very happy'
    
# Create a column named as "happiness_degree".
happiness_df['happiness_degree'] = happiness_df.apply(lambda column: happiness_degree_finder(column), axis=1)

# Print the top 10 records of the updated DataFrame.
happiness_df.head(10)
```

Question 9) Plot multiple bar charts in a single chart that show the mean values of 'Social support','Freedom','Generosity' according to happiness degree (15 points)

```python
# Group data according to "happiness_degree", then find mean of each column.
grouped_df = happiness_df.groupby('happiness_degree')
meanValues = grouped_df.mean()

# Re-index groups from very happy to very unhappy, then reset indexes for use DataFrame.
meanValues = meanValues.reindex(['very happy', 'happy', 'moderately happy', 'unhappy', 'very unhappy'])
meanValues = meanValues.reset_index()

# Determine labels and count of "happiness_degree" types.
labels = meanValues['happiness_degree'].values.tolist()
n = np.arange(len(meanValues))

# Determine the bar chart properties.
socialSupport = meanValues['Social support'].values.tolist()
bar1 = plt.bar(n, socialSupport, 0.25, color='r')
freedom = meanValues['Freedom'].values.tolist()
bar2 = plt.bar(n + 0.25, freedom, 0.25, color='g')
generosity = meanValues['Generosity'].values.tolist()
bar3 = plt.bar(n + 0.50, generosity, 0.25, color='b')

# Determine the x-axis, y-axis and title name.
plt.xlabel('happiness_degree')
plt.ylabel('Mean Value')
plt.title('Mean Values according to happiness_degree')

# Determine the "happiness_degree" values writing rotation and size.
plt.xticks(n + 0.25, labels, rotation='vertical')

# Create a legend for bar charts.
plt.legend((bar1, bar2, bar3), ('Social support', 'Freedom', 'Generosity'))

# Show chart.
plt.show()
```

Question 10) Draw random samples 30%, 40%, 50%, 60% of the data with replacement and investigate how the correlation of sample to the whole data changes with respect to sample size. (15 points)

```python
# Draw the %30 Sample of the DataFrame, then find the Correlation Matrix of %30 Sample.
thirtyPercent = happiness_df.sample(frac=0.30, replace=True, random_state=1)
thirtyCorr = thirtyPercent.corr()

# Show the Correlation Matrix of the %30 Sample.
display(thirtyCorr)
```

```python
# Draw the %40 Sample of the DataFrame, then find the Correlation Matrix of %40 Sample.
fourtyPercent = happiness_df.sample(frac=0.40, replace=True, random_state=1)
fourtyCorr = fourtyPercent.corr()

# Show the Correlation Matrix of the %40 Sample.
display(fourtyCorr)
```

```python
# Draw the %50 Sample of the DataFrame, then find the Correlation Matrix of %50 Sample.
fiftyPercent = happiness_df.sample(frac=0.50, replace=True, random_state=1)
fiftyCorr = fiftyPercent.corr()

# Show the Correlation Matrix of the %50 Sample.
display(fiftyCorr)
```

```python
# Draw the %60 Sample of the DataFrame, then find the Correlation Matrix.
sixtyPercent = happiness_df.sample(frac=0.60, replace=True, random_state=1)
sixtyCorr = sixtyPercent.corr()

# Show the Correlation Matrix of %60 Sample.
display(sixtyCorr)
```

```python
print("""
As a conclusion, while the sample size is increasing, correlation of the sample approaches the correlation of the population.
I mean by this sentence, bigger sample size gives more close correlation values to real correlation of the all data.
According to these hypothesis, choosing bigger sized sample makes easier to determine the real correlation between values.
""")
```

Question 11) Plot pie charts for random sampled data obtained in previous question according to happiness_degree. (10 points)

```python
def pieChartPlotter(df, chartTitle):
    dataframe = df.groupby('happiness_degree')
    degreeCounts = dataframe['happiness_degree'].agg('count')
    chart = degreeCounts.plot(figsize=(7,7), kind='pie', ylabel="", xlabel="", title=chartTitle, autopct='%1.2f%%')
```

```python
# Pie Chart of %30 Sample according to happiness_degree.
pieChartPlotter(thirtyPercent, "%30 Sample according to happiness_degree")
```

```python
# Pie Chart of %40 Sample according to happiness_degree.
pieChartPlotter(fourtyPercent, "%40 Sample according to happiness_degree")
```

```python
# Pie Chart of %50 Sample according to happiness_degree.
pieChartPlotter(fiftyPercent, "%50 Sample according to happiness_degree")
```

```python
# Pie Chart of %60 Sample according to happiness_degree.
pieChartPlotter(sixtyPercent, "%60 Sample according to happiness_degree")
```
