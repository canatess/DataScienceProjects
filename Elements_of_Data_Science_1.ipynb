Question 1) Please read the "cars.csv" file as a DataFrame object. Print the length and the first 5 tuples of the DataFrame.

Question 2) Use "price" attribute to crate array and print mean, minimum, maximum value of price


```python
#Create an array with "price" attribute.
priceArray = cars['price'].to_numpy()

#Print the mean, minimum and maximum value of "price".
print("Mean Value: ")
print(np.mean(priceArray))
print("\nMinimum Value: ")
print(np.amin(priceArray))
print("\nMaximum Value: ")
print(np.amax(priceArray))
```

Question 5) Please read the "Hotels1.csv" and "Hotels2.csv" files as a DataFrame object. Print the length and the first 5 tuples of them.


```python
#Read the files as a DataFrame objects.
hotelOne = pd.read_csv("hotels1.csv")
hotelTwo = pd.read_csv("hotels2.csv")

#Print the length of DataFrames.
print("Length(row, column) of the first hotel DataFrame: ", end='')
print(hotelOne.shape)
print("Length(row, column) of the second hotel DataFrame: ", end='')
print(hotelTwo.shape)

#Print the first 5 tuples of the DataFrames.
display(hotelOne.head(), hotelTwo.head())
```

Question 6)Combine information of these two dataset into a single dataframe without any redundant column and display 10 records of the new dataframe. Be careful combining these two datasets  the new dataframe has a redundant column, find that column and not include

```python
#Combine the DataFrames into a single DataFrame.
combineHotels = pd.merge(hotelOne, hotelTwo, left_index=True, right_index=True)

#Drop the redundant column "name" which informatin has already covered by "hotelname".
hotels = combineHotels.drop(columns = ['name'])

#Show the first 10 record of the new DataFrame.
hotels.head(10)
```

Question 7)Display reviews rating mean by province

```python
#Group the data according to "province", then find mean value of each province.
groupedData = hotels.groupby('province')
meanProvince = groupedData.mean()
meanProvince = meanProvince.reset_index()

#Show the mean review rating of each province.
meanProvince[['province', 'reviews.rating']]
```

Question 8) Write records stored in a DataFrame that is given to you as image to a SQLite database and read all data from SQL via a SQL query


```python
#Create DataFrame.
data = {'index': [0,1,2,3,4,5,6,7],
        'course_codes': ['AIN101', 'AIN212', 'AIN212', 'AIN311', 'AIN440', 'BBM201', 'BBM301', 'BBM301'],
        'course_name': ['Freshman Seminar', 'Elements of Data Science', 'Elements of Data Science', 
                        'Foundations of Machine Learning', 'Introduction to Natural Language Processing', 
                        'Data Structures', 'Programming Languages', 'Programming Languages'],
        'day': ['Monday', 'Tuesday', 'Monday', 'Wednesday', 'Tuesday', 'Friday', 'Thursday', 'Thursday'],
        'number_of_students': [80,35,35,40,25,43,70,70]   
       }

df = pd.DataFrame(data, columns=['index', 'course_codes', 'course_name', 'day', 'number_of_students'])

#Create the database in RAM
conn = connect(':memory:') 

#Create a cursor.
cursor = conn.cursor()

#Create a table.
cursor.execute("""CREATE TABLE Schedule (
              'index' INTEGER PRIMARY KEY,
              course_codes TEXT,
              course_name TEXT,
              day TEXT,
              number_of_students INTEGER)""")

#Write DataFrame to DataBase.
df.to_sql('Schedule', conn, if_exists='replace', index=False)

#Read the data with SQL Query.
schedule = pd.read_sql("SELECT * From Schedule", conn)
schedule
```

Question 9) Filter courses that are hold on Tuesday and number of students are higher than 30 by using a SQL query

```python
#Filter the courses according to "hold on Tuesday and number of students are higher than 30".
filterCourses = pd.read_sql("SELECT * FROM Schedule WHERE day='Tuesday' AND number_of_students > 30", conn)
filterCourses
```

Question 10) Show the total number of students by using a SQL query

```python
#Show the total number of students.
studentCount = pd.read_sql("SELECT SUM(number_of_students) FROM Schedule", conn)
studentCount
```

Question 11) Show course names in ascending order by using a SQL query

```python
#Sort courses according to course name.
sortedCourse = pd.read_sql("SELECT course_name FROM Schedule ORDER BY course_name ASC", conn)
sortedCourse
```

Question 12) Show courses which codes 'AIN 212' by using a SQL query

```python
#Show courses which course code is "AIN212".
courseCode = pd.read_sql("SELECT * FROM Schedule WHERE course_codes='AIN212'", conn)
courseCode
```

Question 13)  Show all data without duplicates by using a SQL query

```python
#Show all data without duplicates.
notDuplicate = pd.read_sql("SELECT DISTINCT [course_codes], [course_name], [day], [number_of_students] FROM Schedule", conn)
notDuplicate
```

Question 14) Change AIN311 course's day to Friday and display it by using a SQL query

```python
#Change AIN311 course day to Friday.
cursor.execute("UPDATE Schedule SET day='Friday' WHERE course_codes='AIN311'")
conn.commit()

#Show updated AIN311.
updatedCourse = pd.read_sql("SELECT * FROM Schedule WHERE course_codes='AIN311'", conn)
updatedCourse
```
