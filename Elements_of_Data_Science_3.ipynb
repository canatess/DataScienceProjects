Question 1) Please read the "sales.csv" file as a DataFrame object named as "sales_df". Print the length and the first 5 tuples of the DataFrame. (5 points)

```python
# Read the "sales.csv" file as a DataFrame object named as "sales_df".
sales_df = pd.read_csv("sales.csv")

# Print the length of the DataFrame.
print("Length(row, column) of the DataFrame: ", end = '')
print(sales_df.shape)

# Print the first 5 tuples of the DataFrame.
sales_df.head()
```

Question2) Split the dataset into the independent (name it 'x') and dependent variables (name it 'y') which are predictors and response variable( 'Item_Outlet_Sales'). Print their shape info and top 10 records of them. (5 points)

```python
# Split the dataset into the independent(x) variables and dependent(y) variable.
x = sales_df.loc[:, sales_df.columns != 'Item_Outlet_Sales']   # Predictors
y = sales_df.loc[:, sales_df.columns == 'Item_Outlet_Sales']   # Response

# Print independent variables shape info and top 10 record.
print("Predictors Shape(row, column) Info: ", end = '')
print(x.shape)
display(x.head(10))

# Print dependent variable shape info and top 10 record.
print("\n\nResponse Shape(row, column) Info: ", end = '')
print(y.shape)
display(y.head(10))
```

Question3) Normalization step: Rescale independent variables (x variable) into the range [0,1] by applying  MinMaxScaler to normalize. Lastly convert into dataframe (10 points)

```python
# Importing MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Define scaler for re-scale independent variables.
min_max_scaler = MinMaxScaler(feature_range = (0,1))

# Re-scale independent variables.
scaled_X = min_max_scaler.fit_transform(x)

# Convert the re-scaled data into DataFrame.
scaled_X = pd.DataFrame(scaled_X, index=x.index, columns = x.columns)

# Show the scaled DataFrame.
scaled_X
```

Question 4) Split data into train and test sets. Use the split ratio as 80:20 and set the 'random_state' to an integer value. (10 points)

```python
# Importing Train test split
from sklearn.model_selection import train_test_split
```

```python
# Split data into train(%80) and test(%20) sets, then set the "random_state" as 1.
x_train, x_test, y_train, y_test = train_test_split(scaled_X, y, train_size = 0.80, random_state = 1)

# Display the train and test sets shape. (EXTRA PART)
display(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

Question 5) Create an instance of kNN with a number of neighbors=3, then fit kNN regression model to train data. Finally, make prediction and calculate mean squared error (MSE). Print MSE value. (15 points)

```python
#importing KNN regressor and metric mse
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as mse
```

```python
# Create an instance of kNN with 3 neighbors.
kNN_instance = KNN(n_neighbors = 3)

# Fit the kNN regression model to train data.
kNN_instance.fit(x_train, y_train)

# Make prediction with fitted model.
test_pred = kNN_instance.predict(x_test)

# Calculate mean squared error, then print it.
mse_value = mse(y_test, test_pred)
print("Mean Squared Error: ", end = '')
print(mse_value)
```

Question 6) Create Elbow function which does the same operation in the previous question for each k instance in the range [1-40] by using Manhattan and Euclidean distance metrics separately in order to determine the optimal value of k. (25 points)

```python
# Create Elbow Function.
def Elbow(rangeStart, rangeEnd, x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.80, random_state = 1)
    global mCoefDict, eCoefDict;
    mCoefDict = {}
    eCoefDict = {}
    
    for k in range(rangeStart, rangeEnd + 1, 1):
        knnModel1 = KNN(k, metric = 'manhattan')
        knnModel2 = KNN(k, metric = 'euclidean')
        
        knnModel1.fit(x_train, y_train)
        knnModel2.fit(x_train, y_train)
        
        knnPredicted1 = knnModel1.predict(x_test)
        manhattan_mse = mse(y_test, knnPredicted1)
        
        knnPredicted2 = knnModel2.predict(x_test)
        euclidean_mse = mse(y_test, knnPredicted2)
        
        mCoefDict[k] = manhattan_mse
        eCoefDict[k] = euclidean_mse
    
    man_min_key = min(mCoefDict.keys(), key = (lambda a: mCoefDict[a]))
    euc_min_key = min(eCoefDict.keys(), key = (lambda b: eCoefDict[b]))
    
    if (mCoefDict[man_min_key] < eCoefDict[euc_min_key]):
        print("Optimal value of k is calculated with Manhattan: ", end = '')
        print(man_min_key)
    elif (mCoefDict[man_min_key] == eCoefDict[euc_min_key]):
        print("Minimum MSE is same for Manhattan and Euclidean metrics.")
        print("\nOptimal Value of k with Euclidean Metric: ", end = '')
        print(euc_min_key)
        print("\nOptimal Value of k with Manhattan Metric: ", end = '')
        print(man_min_key)
    else:
        print("Optimal value of k is calculated with Euclidean: ", end = '')
        print(euc_min_key)
```

Call the defined function for defined range of k

Question 7) Visualize the test mean squared error (MSE) results of Manhattan and Euclidean distance metric by plotting. Two curves should be displayed in the same graph. (15 points)

```python
# Take values and keys from distances dictionaries for create graphs.
man_values = mCoefDict.values()
man_keys = mCoefDict.keys()

euc_values = eCoefDict.values()
euc_keys = eCoefDict.keys()

# Plot lines.
plt.plot(man_keys, man_values, label = 'MSE for Manhattan', color = 'b')
plt.plot(euc_keys, euc_values, label = 'MSE for Euclidean', color = 'r')

# Determine label names and title of graph.
plt.xlabel('k Values')
plt.ylabel('MSE Values')
plt.title('MSE with k Values')

# Create legend.
plt.legend()

# Show the graph.
plt.show() 
```
